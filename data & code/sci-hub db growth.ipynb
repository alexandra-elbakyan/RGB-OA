# Sci-Hub database accumulation over time

This notebook contains code in **Julia language** to plot a timeline graph that shows the number of articles (DOI) in Sci-Hub database against the number of articles indexed in Crossref database from 2014 to 2023. Only DOI of *journal-article* and *proceedings-article* are counted.

The script requres two tabulated files as an input:

1. *crossref date count.tab*: the file contains the number of journal / conferences DOIs that were created in Crossref every day in the format YYYY-MM-DD
2. *sci-hub date count.tab*: same but only counting DOIs that were available in Sci-Hub full-text database.

To create these tabulated files from original Crossref and Sci-Hub indexes, see *preprocessing* folder.

The figure is prepared for the paper 'From black open acces to open access of colour: accepting the diversity of approaches towards free science'.

```julia
using Dates,
      Plots,
      Format,
      Random,
      DelimitedFiles,
      DataStructures

dir = @__DIR__ ;
```

## Loading the data

The counts for Sci-Hub and Crossref are loaded from two tab-delimited files located in the **data** directory:

```julia
function counts(source)
    data = readdlm("$dir/data/$source date count.tab", '\t', String, '\n', quotes = false, use_mmap = true)
    SortedDict(data[:,1] .=> parse.(Int64, data[:,2]))
end

sci_hub, crossref = 

counts("sci-hub"),
counts("crossref") ;
```

## Preparing the data

Article counts are summed up to get cummulative database size for every date:

```julia
function cummulative(counts, start = "2014-01-01", stop = "2023-01-01")

    total = 0
    result = OrderedDict()
    start, stop = Date(start), Date(stop)
    
    date = Date(first(keys(counts)))
    
    while date < stop

        d = string(date)
        n = d in keys(counts) ? counts[d] : 0
        
        total += n

        if date >= start
            result[d] = total
        end
    
        date += Day(1)
    end
    
    result
end

scihub_grow, crossref_grow =

cummulative(sci_hub),
cummulative(crossref) ;
```

## Plotting the data

```julia
x  = collect(0:(length(scihub_grow)-1))
y  = collect(values(scihub_grow))
yy = collect(values(crossref_grow))

ticks = []
labels = []
year = ""
i = 1
for d in keys(scihub_grow)
    if d[1:4] != year
        year = d[1:4]
        push!(ticks, i)
        push!(labels, year)
    end
    i += 1
end
push!(ticks, i)
push!(labels, "2023 ")

yticks = collect(30:10:100) 
ylabels = [string(y, 'm') for y in yticks]
yticks = yticks .* 1000000

colors    = [:lightgray "#C41E3A"]
linecols  = [:white :white]
hatches   = [:x, nothing]
legend    = ["articles indexed in Crossref",
             "articles available on Sci-Hub"]

ft = "sans-serif"

p = plot(xtickfontsize = 8, xtickfontfamily = ft,
         ytickfontsize = 8, ytickfontfamily = ft,
         x_foreground_color_axis = :lightgray,
         x_foreground_color_border = :lightgray,
         y_foreground_color_axis = "#aaa",
         y_foreground_color_border = "#aaa"
)

for (i, yi) in enumerate([yy, y])
    areaplot!(x, yi, c = colors[i], fa = 1, st = :steppost, stack = false,
            title = "",
            label = legend[i],
            foreground_color_legend = "#aaa",
            legendfontfamily = ft,
            legendfontsize = 9,
            xticks = (ticks, labels), yticks = (yticks, ylabels),
            size = (512 * 2, 512),
            linecolor = linecols[i],
            linewidth = 4, dpi = 300, fillstyle = hatches[i],
            widen=false,tickdirection=:out
    )
end

for yi in [2014, 2016, 2018, 2020]

    i = first(findall(j -> j == string(yi), labels))
    ti = ticks[i]

    if ti > 1
        plot!([ti, ti], [0, y[ti] - 4 * 10^6], linecolor = :white, linewidth = 1, linestyle = :dash, label = nothing)
    end
        
    percent = Int64(round(y[ti] * 100 / yy[ti]))
    n = string(yi, ": ", "$percent%\n", cfmt("%\'d", y[ti]))

    annotate!(ti + 64, y[ti] - 6 * 10^6, text(n, :white, 10, halign = :left, family = ft))

    plot!([ti], [y[ti] - 6 * 10^6], markershape = :circle, markerstrokecolor = :white, markercolor = :white, markersize = 10, label = nothing)
    
end

p
```
