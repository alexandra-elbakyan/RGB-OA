# Sci-Hub monthly download stats

This notebook contains Python code to plot a figure representing total number of Sci-Hub PDF downloads worldwide each month, for the paper 'From black open access to open access of colour: accepting the diversity of approaches towards free science'.

The script uses .json monthly stats provided by Sci-Hub.

```python
import os
import json
import matplotlib.pyplot as pyplot
import matplotlib.patches as mpatch

from glob import glob
from tqdm import tqdm
from matplotlib.ticker import FuncFormatter
from matplotlib.patches import FancyBboxPatch
```

## Loading input data

The input file contains number of downloads and users for every month since 2016, in the following format:

- **users** number of users as calculated by unique IP address
- **users:cuts** number of users identified as bots
- **downloads** number of PDF downloads
- **downloads:cut** number of PDF downloads from users identified as bots

```python
months = json.loads (open("data/sci-hub monthly stats.json").read())
months
```

The data is prepared for plotting by excluding downloads coming from robots:

```python
downloads = {}
for month, numbers in months.items():
    downloads[month] = 0
    if 'downloads' in numbers:
        downloads[month] = numbers['downloads'] - numbers['downloads:cut']
```

Function to plot the data:

```python
def monthview(downloads, # dictionary containing number of downloads for every month as an input
              figsize = (32, 24),
              palette = ['#b367a1', '#ec7793', '#ff977a', '#ffc567', '#92ea84', '#0095ce', '#a2acbd', '#007f3a', '#b94916'],
              save_to = 'figures/sci-hub downloads month.png', dpi = 150):
    
    mi = list(downloads.keys())
    ms = int(mi[0][0:4])

    # prepare different color for each year according to the palette
    colors = [palette[(int(month[0:4])-ms)%len(palette)] for month in mi]

    # prepare the canvas for drawing
    fig, ax = pyplot.subplots(1, 1, figsize = figsize)
    pyplot.grid(axis = 'y', color = 'grey', linestyle = '-.', linewidth = 1, zorder = 0)

    # plot the bar for each month in the data
    bars = pyplot.bar(*zip(*downloads.items()), width = 1, zorder = 2)
    heights = [bar.get_height() for bar in bars]

    # find month with largest number of downloads for every year
    ny = len({month[0:4] : True for month in mi})
    th = [max(heights[i: i + 12]) for i in range(0, ny * 12, 12)]
    mh = max(th)
    ah = sum(th)/len(th)
    
    ph = 0
    hy = ah//2
    yticks = []
    numbers = []

    # iterate through all the months to do the following:

    # - color every year according to the palette
    # - draw bar borders
    # - draw lines between different years
    # - highlight summer months for easier perception
    # - annotate months with largest download counts
    # - annotate different years
    
    for i, bar in enumerate(bars):
        
        x, y = bar.get_xy()
        w, h = bar.get_width(), bar.get_height()

        # color the bar according to the palette
        bar.set_color(colors[i])
        bar.set_alpha(1)

        # highlight summer month
        if mi[i][6] in ['6','7','8']:
            bar.set_alpha(0.8)
            bar.set_hatch('x')
            bar.set_edgecolor('white')

        bar_border = 'black'
        bar_border_w = 3
        
        if i > 1:

            # if the year has changed, draw a separator line
            if mi[i][0:4] != mi[i-1][0:4]:
                ax.plot([x, x], [y, y + h], color = bar_border, lw = bar_border_w, zorder = 4)

            # annotate the current year
            elif i > 2:
                if mi[i][0:4] != mi[i-2][0:4]:
                    
                    ax.text(x, 3000000, mi[i][0:4],
                            fontsize = 28,
                            bbox = dict(boxstyle = "round", fc = "w", ec = "w"),
                            backgroundcolor = 'white')

        # if that is the last bar, draw border line
        if i == len(mi) - 1:
            ax.plot([x + w, x + w], [y, y + h], color = bar_border, lw = bar_border_w)

        # draw border on the top of each bar
        ax.plot([x, x + w], [y + h, y + h], color = bar_border, lw = bar_border_w)

        # draw side borders for each bar
        if h > ph:
            ax.plot([x, x], [ph, h], color = bar_border, lw = bar_border_w)
        if h < ph:
            ax.plot([x, x], [h, ph], color = bar_border, lw = bar_border_w)
            
        ph = h

        # annotate months with largest number of downloads
        for hi in th:
            if h == hi:
                n = '{:,}'.format(h)
                nx = x - 3 * w - len(n) * 0.9
                ny = (h + mh * 0.04) * 0.99
                overlap = False
                for ni, nyi in numbers:
                    if i-ni < 10 and abs(ny/mh-nyi)*10 < 0.5:
                        overlap = True
                if not overlap:
                    ax.plot([x - 3 * w, x], [h + mh * 0.04, h + mh * 0.04], color = "black", lw = 2, linestyle = '--')
                    ax.plot([x, x], [h , h + mh * 0.04], color = "black", lw = 2, linestyle = '--')
                    ax.text(nx, ny, n, fontsize = 26,
                                       color = 'black',
                                       fontweight = 'normal',
                                       bbox = dict(boxstyle = "round", fc = "w", ec = "black"),
                                       backgroundcolor = 'white')
                    numbers.append((i, ny/mh))
                    break
    
        yticks.append(h)

    # y-axis will have 8 ticks between min and max annotated
    yma = max(yticks)
    ymi = min(yticks)
    ys = (yma-ymi)//8
    yticks = [yti for i, yti in enumerate(range(ymi,yma,ys)) if i > 0]
    
    ax.set_yticks(yticks) ;
    ax.set_xticks([]) ;
    
    ax.tick_params(axis='y', which = 'major', labelsize = 32)
    ax.ticklabel_format(axis = 'y', style = 'plain', useOffset = False)
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: str(round(y/10**6,1)) + 'M' if y > 10**6 else str(y//10**3) + 'K'))

    fig.savefig(save_to, dpi = dpi)
```

```python
monthview(downloads)
```
