# Sci-Hub article availability per publisher

This notebook contains code in **Julia language** to plot bar charts showing percentage of documents (DOI) available in Sci-Hub, for different publishers and year of publication.

The script requres preprocessed data to be available in **data** folder. The folder should contain two sub-folders, *publishers* and *publishers Sci-Hub* that contain a separate .txt file for every publisher indexed in Crossref. The file contains number of articles published every year as by Crossref index, and the number of articles published the same years that had PDF full-text available in Sci-Hub.

To create preprocessed data from original Crossref and Sci-Hub indexes, see *preprocessing* folder.

The figures are prepared for the paper 'From black open access to open access of color: accepting the diversity of approaches towards free science'.

```julia
using Glob,
      Plots,
      CairoMakie,
      Statistics,
      ProgressMeter,
      DelimitedFiles,
      DataStructures,
      OrderedCollections

      dir = @__DIR__ ;
```

The percentage of articles available in Sci-Hub will be calculated within the following timespans:

```julia
a..b = (a:b-1)

dates = [ 0..2012, 2012..2016, 2016..2020,  2020..2024 ] ;
```

## Loading the data

The following function is used to load data for every publisher to a dictionary:

```julia
function yearcounts(hashes, source = "publishers")

    result = Dict()
    
    @showprogress output = stdout for hash in hashes
        txt = string(dir, "/data/", source, "/", hash, ".txt")
        
        publisher = open(txt) do i readline(i) end
        counts = readdlm(txt, '\t', String, '\n', quotes = false, use_mmap = true, skipstart = 1)
        counts = SortedDict(parse.(Int, counts[:,1]) .=> parse.(Int, counts[:,2]))

        result[publisher] = counts
    end

    result
end ;
```

The publisher yearly count data is loaded both for Sci-Hub and Crossref:

```julia
hashes = glob("*.txt", dir * "/data/publishers sci-hub")
hashes = [split(h, "/")[end][1:end-4] for h in hashes]

crossref_counts = yearcounts(hashes)
scihub_counts   = yearcounts(hashes, "publishers sci-hub")
```

## Processing the data

The following function will calculate the percentage of articles available in Sci-Hub, for selected publishers and timespans of interest:

```julia
function calculate(scihub, crossref, selection, dates)

    results = Dict("percentage" => Dict(),
                   "sci-hub"    => Dict(),
                   "crossref"   => Dict())
    
    @showprogress output = stdout for pub in selection
    
        for i in keys(results)
            results[i][pub] = Dict() end
        
        for span in dates
            k = 0
            n = 0
            
            for year in span
                if year in keys(scihub[pub])
                    k += scihub[pub][year]
                    n += crossref[pub][year]
                end
            end
    
            if n > 0
                span = string(span[1],"-", span[end] + 1)
                results["sci-hub"][pub][span]  = k
                results["crossref"][pub][span] = n
                results["percentage"][pub][span] = round(k * 100 / n)
            end
        end
    
        for i in keys(results)
            results[i][pub] = sort(results[i][pub]) end
        
    end

    results

end
```

Run the calculation for selected publishers and timespans and display the result for *Wiley* :

```julia
selected = Dict("Elsevier BV" => "Elsevier",
                "Wiley"       => "Wiley",
                "Springer Science and Business Media LLC" => "Springer",
                "IEEE" => "IEEE")

results = calculate(scihub_counts, crossref_counts, keys(selected), dates)

results["percentage"]["Wiley"]
```

## Plotting selected publishers

Function to create bar plot from calculated data:

```julia
function bars(results, selected, dates, palette)
    
    spans, percents, publishers, colors = Integer[], [], Integer[], []
    
    for (s, span) in enumerate(dates)
        span = string(span[1], "-", span[end] + 1)
        
        for (i, publisher) in enumerate(keys(selected))
            push!(spans, s)
            push!(publishers, i)
            push!(percents, results["percentage"][publisher][span])
            push!(colors, palette[i])
        end
    end
    
    data = (spans      = spans,
            percents   = percents,
            publishers = publishers)
    
    fig = Figure(size = (1400, 480) , fontsize = 22)

    ticks = (1:length(dates), string.(dates))
    ticks[2][1] = "before\n2012"
    
    ax = Axis(fig[1,1], xticks = ticks, 
            title = "", # "Percentage of articles available in Sci-Hub",
            limits = (nothing, nothing, 0, 128),
            xgridvisible = false, ygridvisible = false
    )
    
    barplot!(ax, data.spans, data.percents,
            dodge = data.publishers,
            color = colors,
            gap = 0.1,
            dodge_gap = 0,
            bar_labels = :y,
            label_offset = 8,
            label_size = 20,
            strokewidth = 2,
            strokecolor = :white,
            label_formatter = p -> string(trunc(Int, p), "%"),
    )

    hidespines!(ax)
    
    labels = collect(values(selected))
    elements = [PolyElement(polycolor = palette[i]) for (i, p) in enumerate(keys(selected))]
    title = "" # "Publishers"
    
    Legend(fig[1,2],
           elements,
           labels,
           title,
           framevisible = false,
           labelfontfamily = "sans-serif")

    fig

end
```

```julia
palette = ["#0095ce", "#00b7c6", "#00d4a8", "#92ea84"]
```

```julia
fig = bars(results, selected, dates, palette)
```

```julia
save("$dir/figures/sci-hub percent publisher (a).png", fig)
```

## Plotting top N publishers

First we need to create a publisher top-chart, according to the number of articles indexed in Crossref, with largest publishers on the top:

```julia
publishers = glob("*.txt", dir * "/data/publishers")

pubtop = OrderedDict()

@showprogress output = stdout (
for txt in publishers
    publisher = open(txt) do i readline(i) end
    counts = readdlm(txt, '\t', String, '\n', quotes = false, use_mmap = true, skipstart = 1)
    counts = Dict(counts[:,1] .=> parse.(Int, counts[:,2]))
    pubtop[publisher] = sum(values(counts))
end )

pubtop = sort(pubtop; byvalue = true, rev = true)
pubtop
```

For selected top K publishers, calculate the total (summed) number of articles published every year:

```julia
function groupcounts(counts, pubtop, top_k = [8, 64, 512])

    counts["all"] = OrderedDict()
    for t in top_k
        counts["top $t"] = OrderedDict() end
    
    @showprogress output = stdout (
    for (i, publisher) in enumerate(pubtop)
        publisher ∉ keys(counts) &&
            continue
            
        for year in keys(counts[publisher])

            for t in top_k
                year ∉ keys(counts["top $t"]) &&
                           (counts["top $t"][year] = 0)

                i <= t &&  (counts["top $t"][year] += counts[publisher][year])
            end
    
            year ∉ keys(counts["all"]) &&
                (counts["all"][year]  = 0)
                 counts["all"][year] += counts[publisher][year]
            
        end
        
    end )
    
    counts
end
```

```julia
scihub_counts_g = groupcounts(scihub_counts, keys(pubtop))
crossref_counts_g = groupcounts(crossref_counts, keys(pubtop)) ;
```

Plot bar charts using cummulative counts for top K publishers:

```julia
grouped = OrderedDict("top 8"    => "top 8",
                      "top 64"   => "top 64",
                      "top 512"  => "top 512",
                      "all"      => "all publishers") ;

results = calculate(scihub_counts_g, crossref_counts_g, keys(grouped), dates)
fig = bars(results, grouped, dates, palette)
```

```julia
save("$dir/figures/sci-hub percent publisher (b).png", fig)
```
