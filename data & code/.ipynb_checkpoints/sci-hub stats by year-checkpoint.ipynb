```python
import os
import json
import random
import pandas
import seaborn
import matplotlib.pyplot as pyplot
import matplotlib.style as style

from glob import glob
from tqdm.notebook import tqdm
```

```python
months = json.loads(open('data/sci-hub monthly stats.json').read())
months
```

```python
totals = {'users' : {}, 'pdfs' : {}, 'users:full' : {}, 'pdfs:full' : {}}

for year in range(2016, 2024):
    for k in totals.keys():
        totals[k] [year] = 0
    for month in range(1,12 +1):
        if month < 10:
            month = '0' + str(month)
        ym = f'{year}-{month}'
        if ym in months and (ym not in ['2018-10', '2018-11']):
            totals['users'][year] += months[ym]['users'] - months[ym]['users:cut']
            totals['users:full'][year] += months[ym]['users']
            totals['pdfs'][year] += months[ym]['downloads'] - months[ym]['downloads:cut']
            totals['pdfs:full'][year] += months[ym]['downloads']

totals
```

```python
downloads = [{'year'      : year,
              'users'     : totals['users'][year],
              'users:full': totals['users:full'][year],
              'pdfs'      : totals['pdfs'][year],
              'pdfs:full' : totals['pdfs:full'][year]}
             
             for year in sorted(totals['users'].keys())]

data = pandas.DataFrame(downloads)
data.set_index('year', inplace = True)
data
```

```python
pyplot.style.use('classic')

fig = pyplot.figure(figsize = (18, 10), facecolor = 'white')

pyplot.xticks([])
pyplot.yticks([])

colors = ['darkred', 'darkcyan']
months = [str(y) for y in sorted(totals['pdfs'].keys())]

for i, (value, legend, title) in enumerate([('pdfs', ['number of PDF downloads', 'number of downloads by bots'], 'number of pdf downloads from Sci-Hub every year'),
                                            ('users', ['number of users accessing Sci-Hub'], 'number of users accessing Sci-Hub every year\n(estimated by unique IP addresses)')]):
    ax = fig.add_subplot(121 + i)
    ax.ticklabel_format(style = 'plain')
#    ax.set_facecolor("white")
    
    data[value].plot(ax = ax, kind = "barh", rot = 0, width = 0.98, color = colors[i], edgecolor = 'white', zorder = 2)

    for p in ax.patches:
        label = '{:,}'.format(p.get_width())
        ax.annotate(label, (0, p.get_y() + (p.get_height() * 0.5)),
                    backgroundcolor = '1',
                    textcoords = 'offset points',
                    xytext = (16, 0)
                   )

    data[value + ':full'].plot(ax = ax, kind = "barh", rot = 0, width = 0.98, color = 'white', hatch = '///', edgecolor = 'gray', zorder = 1)
    
    ax.set(yticklabels = months, xticklabels = [], ylabel = None, xlabel = None)
    
    for spine in ax.spines.values():
        spine.set_edgecolor('lightgray')

    pyplot.savefig('figures/sci-hub downloads users year (' + chr(ord('a') + i) + ').png')
```
